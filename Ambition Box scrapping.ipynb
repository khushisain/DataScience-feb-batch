```python
import requests

import bs4
```

```python
company_names = []
company_ratings = []
company_details = []


url = "https://www.ambitionbox.com/list-of-companies?"

for i in range(1, 201):
    
    dict1 = {
        'sortBy':'popular',
        'locations':'jaipur',
        'page':i
    }

    header = {'user-agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"}
    
    resp = requests.get(url, params=dict1, headers=header)
    if resp.status_code == 200:
        soup = bs4.BeautifulSoup(resp.content)
        all_divs = soup.find_all('div', attrs={'class':"companyCardWrapper__metaInformation"})
        
        for i in all_divs:
            name = i.h2.text.strip()
            rating = i.find('span', attrs={'class':"companyCardWrapper__companyRatingValue"}).text    
            details = i.find('span', attrs={'class':"companyCardWrapper__interLinking"}).text.strip()    
            company_names.append(name)
            company_ratings.append(rating)
            company_details.append(details)
```
